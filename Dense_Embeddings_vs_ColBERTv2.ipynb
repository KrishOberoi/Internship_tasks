```python
from google.colab import files
from llama_index.readers.file import PDFReader

# Step 1: Upload multiple PDF files
uploaded = files.upload()  # allows you to select and upload multiple files

# Step 2: Initialize PDFReader
loader = PDFReader()

# Step 3: Load data from all uploaded PDFs
documents = []
for filename in uploaded.keys():
    print(f"Loading {filename}...")
    docs = loader.load_data(filename)
    documents.extend(docs)

print(f"✅ Loaded {len(documents)} documents from {len(uploaded)} PDFs.")
```

```python
list_pdf_documents = [document.text for document in documents]
```

```python
from ragatouille import RAGPretrainedModel

RAG = RAGPretrainedModel.from_pretrained("colbert-ir/colbertv2.0")
```

```python
RAG.index(
    collection=list_pdf_documents,
    index_name="colbert",
    max_document_length=256,
    split_documents=True,
)
```

```python
results = RAG.search(query="who is kitty? ", k=3, index_name="colbert")
```

```python
results
```

```python
for i, doc, in enumerate(results):
    print(f"---------------------------------- doc-{i} ------------------------------------")
    print(doc["content"])
```

```python
from langchain.vectorstores import FAISS
```

```python
from sentence_transformers import SentenceTransformer

# Load a MiniLM model (this one is commonly used for semantic search)
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
from langchain.vectorstores import FAISS
from langchain.embeddings.base import Embeddings

class MiniLMEmbeddings(Embeddings):
    def __init__(self, model):
        self.model = model

    def embed_documents(self, texts):
        return self.model.encode(texts, normalize_embeddings=True)

    def embed_query(self, text):
        return self.model.encode([text], normalize_embeddings=True)[0]
```

```python
embedding = MiniLMEmbeddings(model)

docsearch = FAISS.from_texts(list_pdf_documents, embedding)
```

```python
similar_docs = docsearch.similarity_search("who is kitty ")
```

```python
for i, doc in enumerate(similar_docs):
  print(f"---------------------------------- doc-{i} ------------------------------------")
  print(doc.page_content)
```
